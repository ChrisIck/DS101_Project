```python
def add_weighted_review_cnt(df,weight):
    weighted_average = [1+(sum(x)*weight) if x != 0 else 1 for x in zip(df['funny'].values, df['cool'].values,df['useful'].values)]
    df['weighted_average'] = weighted_average
    return df
```
