```python
import yelp_functions
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
import ast
```

```python
business = pd.read_json('yelp_academic_dataset_business.json', lines=True)
AZ_reviews = pd.read_csv('AZ_review.csv')
tips = pd.read_json('yelp_academic_dataset_tip.json', lines=True)
```

```python
restaurants = yelp_functions.Clean_Business_Data(business)
restaurants = yelp_functions.enrich_biz_categories(restaurants,.10)
restaurants = yelp_functions.enrich_biz_attributes(restaurants)
restaurants = yelp_functions.join_datasets(AZ_reviews,restaurants,tips)
```

```python
restaurants
```

```python
restaurants = yelp_functions.add_unique_users_per_bid(restaurants)
restaurants = yelp_functions.add_useful_funny_cool_cnt(restaurants)
restaurants = yelp_functions.add_avg_star_rating(restaurants)
restaurants = yelp_functions.add_num_tips_per_bid(restaurants)
restaurants = yelp_functions.add_num_likes_for_tips_per_bid(restaurants)
```
